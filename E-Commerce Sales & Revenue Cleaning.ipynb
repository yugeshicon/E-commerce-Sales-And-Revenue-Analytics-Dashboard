## E-Commerce Sales Data Cleaning

### 1. Import Library

```python
import pandas as pd
```

```python
df = pd.read_csv('E-Commerce Original Dataset.csv')
print("Dataset Loaded Successfully.")
```

### 4. Check Dataset Structure

##### Dataset Shape

```python
print('Dataset Shape:', df.shape)
```

##### Change the Data Type

```python
median = df['Age'].median()
df['Age'] = df['Age'].fillna(median).astype(int)
df['Total Price'] = df['Total Price'].astype(float)
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
print("Data Types Changed Successfully.")
```

##### Dataset Information

```python
df.info()
```

### 5. Generate Summary Statistics

```python
print("Statistical Summary:")
df[['Age', 'Quantity', 'Unit Price', 'Total Price', 'Shipping Fee']].describe()
```

### 6. Detect & Handle Missing Values

```python
df.isnull()
```

```python
df.dropna()
```

```python
df.dropna(subset=['Region', 'Shipping Status'], inplace=True)
```

### 7. Detect & Remove Duplicate Records

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

### 8. Correct Data Types

```python
df['Unit Price'] = df['Unit Price'].round(2)
df['Total Price'] = df['Total Price'].round(2)
df['Shipping Fee'] = df['Shipping Fee'].round(2)
```

### 9. Standardize Text Data

```python
columns = ['Customer ID', 'Gender', 'Region', 'Product Name', 'Category', 'Shipping Status']
df[columns] = df[columns].apply(lambda x: x.str.strip())
df[columns] = df[columns].apply(lambda x: x.str.title())
print("Text Cleaned Successfully.")
```

### 10. Detect & Handle Outliers

##### Outlier Limits Using The IQR Method

```python
num_columns = df.select_dtypes(include='number')

Q1 = num_columns.quantile(0.25)
Q3 = num_columns.quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print("Lower Limit:\n",round(lower,3))
print("\nUpper Limit:\n",round(upper,3))
```

##### Detect Outliers - Head View

```python
outliers = df[(num_columns < lower) | (num_columns > upper)]
print('Detect Outliers: Head Part:')
outliers.head()
```

##### Detect Outliers - Tail View

```python
print('\nDetect Outliers: Tail Part')
outliers.tail()
```

#### Remove Outliers In Dataset

```python
df = df[((num_columns >= lower) & (num_columns <= upper)).all(axis=1)]
df.head()
```
